1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Carregar o DataFrame
df = pd.read_csv('previsao_de_renda (1).csv')

# Verifique as primeiras linhas
print(df.head())
````

````python
print(df.columns)
````

````python
# Contagem de pessoas com posse de imóvel ao longo do tempo
df_imovel = df[df['posse_de_imovel'] == True].groupby('data_ref').size()

# Contagem de pessoas com posse de veículo ao longo do tempo
df_veiculo = df[df['posse_de_veiculo'] == True].groupby('data_ref').size()

# Plot
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
df_imovel.plot(kind='bar', color='skyblue')
plt.title('Distribuição da Posse de Imóvel ao Longo do Tempo')
plt.ylabel('Contagem')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
df_veiculo.plot(kind='bar', color='salmon')
plt.title('Distribuição da Posse de Veículo ao Longo do Tempo')
plt.ylabel('Contagem')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
# Calculando a média de renda para cada sexo ao longo do tempo
media_sexo = df.groupby(['data_ref', 'sexo'])['renda'].mean().unstack()

# Gráfico
media_sexo.plot(figsize=(15, 6))
plt.title('Média de Renda por Sexo ao Longo do Tempo')
plt.ylabel('Média de Renda')
plt.xticks(rotation=45)
plt.show()
````
